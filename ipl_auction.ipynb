```python
# average sold price for each age category by grouping age and then applying mean on sold price

df.groupby('AGE')['SOLD PRICE'].mean()
```

The average price is highest for age category 1 and lowest for category 2 .it shows a pd series to create a datframe we can call reset_index()

```python
# converting the above series into dataframe
sold_price_by_age = df.groupby('AGE')['SOLD PRICE'].mean().reset_index()
print(sold_price_by_age)
```

```python
# average sold price for players for each age and playing role category
sold_price_by_age_role = df.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
print(sold_price_by_age_role)
```

```python
# Comparing avg sold price for differrent age categories with the differrent age and playing role categories .

sold_price_comparission = sold_price_by_age_role.merge(sold_price_by_age, on ='AGE', how = 'outer')
print(sold_price_comparission)
```

```python
# renaming the columns

sold_price_comparission.rename(columns = {'SOLD PRICE_x':'SOLD_PRICE_AGE_ROLE','SOLD PRICE_y':'SOLD_PRICE_AGE'}, inplace = True)
sold_price_comparission.head()
```
